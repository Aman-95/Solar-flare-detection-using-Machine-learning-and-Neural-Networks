#**SVR (rbf kernel)**

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_excel("Dataset_set_2.xlsx")
```

#For c-class solar flares

```
X = dataset.iloc[1:, 0:38].values
y = dataset.iloc[1:, 38].values
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 0)
```

```
from sklearn.preprocessing import StandardScaler
sc_X_train = StandardScaler()
sc_y_train = StandardScaler()
X_train = sc_X_train.fit_transform(X_train)
X_test = sc_X_train.transform(X_test)
#y_train = sc_y_train.fit_transform(y_train)
#y_test = sc_y_train.transform(X_test)
```

```
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
```

```
y_pred = regressor.predict(X_test)
y_test=np.asarray(y_test,dtype=int)
y_pred=np.asarray(y_pred,dtype=int)
```

```
for i in range(0,len(y_pred)):
  if y_pred[i]>=1:
    y_pred[i]=1
  else:
    y_pred[i]=0
  if y_test[i]>=1:
    y_test[i]=1
  else:
    y_test[i]=0
```

```
from sklearn.metrics import confusion_matrix
cm_1 = confusion_matrix(y_test, y_pred)
cm_1
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
